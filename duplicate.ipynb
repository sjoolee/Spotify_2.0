```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import statsmodels.api as sm
import statsmodels.stats.api as sms
import seaborn as sns
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from statsmodels.graphics.tsaplots import plot_acf
import warnings
import csv
import json
import string
import re
from datetime import datetime
```

```python
combined_dupe = 'all_combined_duplicated/Combined_data_dupe.csv'
combined = 'all_combined/Combined_data.csv'
```

```python
df_combined = pd.read_csv(combined)
```

```python
new_rows = []
for i, row in df_combined.iterrows():
    play_count = int(row['Play Count'])
    for j in range(play_count):
        new_rows.append(row)

df_combined_dupe = pd.DataFrame(new_rows)
df_combined_dupe.reset_index()
df_combined_dupe
```

```python
df_combined_dupe.to_csv(combined_dupe, index=False)
```
